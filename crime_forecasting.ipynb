```python
# Dependencies
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
import os
```

```python
# Load data
crime_data = pd.read_csv("sources/yearly_numbers.csv",encoding="utf")
year = crime_data.iloc[:, 1]
```

```python
# Grab number of killed 
n_killed = crime_data.iloc[:, 2]
nk_slope, nk_int, nk_r, nk_p, nk_std_err = stats.linregress(year, n_killed)
nk_fit = nk_slope * year + nk_int
print(f'{nk_fit}')
```

```python
# Grab number of injured
n_injured = crime_data.iloc[:, 3]
ni_slope, ni_int, ni_r, ni_p, ni_std_err = stats.linregress(year, n_injured)
ni_fit = ni_slope * year + ni_int
print(f'{ni_fit}')
```

```python
# Plot
fig, (ax1, ax2) = plt.subplots(2, sharex=True)
fig.suptitle("Gun Incidents Over Time", fontsize=16, fontweight="bold")

ax1.set_xlim(min(year), max(year))
ax1.plot(year, n_killed, linewidth=1, marker="o")
ax1.plot(year, nk_fit, "b--", linewidth=1)
ax1.set_ylabel("Number of killed Rate")

ax2.plot(year, n_injured, linewidth=1, marker="o", color="r")
ax2.plot(year, ni_fit, "r--", linewidth=1)
ax2.set_ylabel("Number of injured Rate")
```

```python
# Print results and save image
year = 2020
print("The number of people kiiled  in 2020 will be " +
      str(nk_slope * year + nk_int) + ".")
print("The numner of people injured in 2020 will be " +
      str(ni_slope * year + ni_int) + ".")

plt.savefig("forecasting_plot.jpg")
```
