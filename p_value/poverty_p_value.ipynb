## P_value 
We decide to calculate p-values of data to check the validity of our claim about correlation between datas.
We want to understand that is there any correlation between poverty and gun incidens ?

```python
p_value_data = county_data[["Number of Incidents",
                            "Poverty Rate"]]
p_value_data.head()
```

```python
poverty_m = p_value_data["Poverty Rate"].mean()
poverty_m
```

```python
poverty_range = p_value_data["Poverty Rate"].unique().tolist()
poverty_range
```

## Good range poverty

```python
range_good = p_value_data.loc[p_value_data["Poverty Rate"] <= 17 ]
range_good.sort_values(by = ["Poverty Rate"], ascending = False)
```

```python
mean_good = range_good["Poverty Rate"].mean()
mean_good
```

## Medium range poverty

```python
range_medium = p_value_data.loc[p_value_data["Poverty Rate"] > 17]
range_medium = range_medium.loc[range_medium["Poverty Rate"] <= 34 ]
range_medium.sort_values(by=["Poverty Rate"], ascending = False)
```

```python
mean_medium = range_medium["Poverty Rate"].mean()
mean_medium
```

## Bad range poverty

```python
range_bad = p_value_data.loc[p_value_data["Poverty Rate"] > 34]
range_bad = range_bad.loc[range_bad["Poverty Rate"] <= 52]
range_bad.sort_values(by = ["Poverty Rate"], ascending = False)
```

```python
mean_bad = range_bad["Poverty Rate"].mean()
mean_bad
```

```python
print(f' The average number of good range is {mean_good}.')
print(f' The average number of medium range is {mean_medium}.')
print(f' The average number of bad range is {mean_bad}.')
```

```python
expected_value = (mean_good + mean_medium + mean_bad / 3)
expected_value
```

```python
observed = pd.Series([mean_good, mean_medium, mean_bad], index = ["Good", "Medium", "Bad"])
observed
```

```python
df = pd.DataFrame([observed]).T
df.head()
```

```python
df[1] = expected_value
```

```python
df.columns = ["Observed", "Expected"]
df
```

```python
# The degree of freedom is 3-1 = 2
# With a p-value of 0.05, the confidence level is 1.00-0.05 = 0.95.
critical_value = stats.chi2.ppf(q = 0.95, df = 2)
critical_value
```

```python
# Run the chi square test with stats.chisquare()
stats.chisquare(df['Observed'], df['Expected'])
```

### There is a significance between poverty and gun violence 
We conclude from p-value which is less than 0.05 we reject the null hypothesis that there is no difference between the means and conclude that a significance correlation between unemployment rate and gun incidents which demonstrates that higher poverty rate will increase gun crimes.

```python
l=[]
trace0= go.Scatter(
        y = county_data['Poverty Rate'],
        x = county_data["Number of Incidents"],
        mode= 'markers',
        name='Poverty (%)',
        marker= dict(size= county_data['Poverty Rate'].values,
                    line= dict(width=1),
                    color= county_data['Poverty Rate'].values,
                    opacity= 0.7,
                    colorscale='Portland',
                    showscale=True),
        text= county_data['State'].values) # The hover text goes here... 
l.append(trace0);

layout= go.Layout(
    title= 'Scatter plot of Poverty rates in 2017',
    hovermode= 'closest',
    xaxis= dict(
        ticklen= 5,
        zeroline= False,
        gridwidth= 2,
    ),
    yaxis=dict(
        title= 'Poverty Rate (%)',
        ticklen= 5,
        gridwidth= 2,
    ),
    showlegend= False,
    )
fig= go.Figure(data=l, layout=layout)
py.iplot(fig)
```
