```python
county_data = pd.DataFrame(resource)
county_data
```

## P_value 
We decide to calculate p-values of data to check the validity of our claim about correlation between datas.
We want to understand that is there any correlation between unemployment and gun incidens ?

```python
p_value_data = county_data[["Number of Incidents",
                            "Unemployment Rate"]]
p_value_data.head()
```

```python
unemployment_m = p_value_data["Unemployment Rate"].mean()
unemployment_m
```

```python
unemployment_range = p_value_data["Unemployment Rate"].unique().tolist()
unemployment_range
```

## Good range unemployment 

```python
range_good = p_value_data.loc[p_value_data["Unemployment Rate"] <= 4 ]
range_good.sort_values(by = ["Unemployment Rate"], ascending = False)
```

```python
mean_good = range_good["Unemployment Rate"].mean()
mean_good
```

## Medium range unemployment

```python
range_medium = p_value_data.loc[p_value_data["Unemployment Rate"] > 4]
range_medium = range_medium.loc[range_medium["Unemployment Rate"] <= 8 ]
range_medium.sort_values(by=["Unemployment Rate"], ascending = False)
```

```python
mean_medium = range_medium["Unemployment Rate"].mean()
mean_medium
```

## Bad range unemployment

```python
range_bad = p_value_data.loc[p_value_data["Unemployment Rate"] > 8]
range_bad = range_bad.loc[range_bad["Unemployment Rate"] <= 13]
range_bad.sort_values(by = ["Unemployment Rate"], ascending = False)
```

```python
mean_bad = range_bad["Unemployment Rate"].mean()
mean_bad
```

```python
print(f' The average number of good range is {mean_good}.')
print(f' The average number of medium range is {mean_medium}.')
print(f' The average number of bad range is {mean_bad}.')
```

```python
expected_value = (mean_good + mean_medium + mean_bad / 3)
expected_value
```

```python
observed = pd.Series([mean_good, mean_medium, mean_bad], index = ["Good", "Medium", "Bad"])
observed
```

```python
df = pd.DataFrame([observed]).T
df.head()
```

```python
df[1] = expected_value
```

```python
df.columns = ["Observed", "Expected"]
df
```

```python
# The degree of freedom is 3-1 = 2
# With a p-value of 0.05, the confidence level is 1.00-0.05 = 0.95.
critical_value = stats.chi2.ppf(q = 0.95, df = 2)
critical_value
```

```python
# Run the chi square test with stats.chisquare()
stats.chisquare(df['Observed'], df['Expected'])
```

### There is a significance between unemployment and gun violence 
We conclude from p-value which is less than 0.05 we reject the null hypothesis that there is no difference between the means and conclude that a significance correlation between unemployment rate and gun incidents which demonstrates that higher unemployment rate will increase gun crimes.
