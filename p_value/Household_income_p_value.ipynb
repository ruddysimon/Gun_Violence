## P_value 
We decide to calculate p-values of data to check the validity of our claim about correlation between datas.
We want to understand that is there any correlation between household income and gun incidens ?

```python
p_value_data = county_data[["Number of Incidents",
                            "Household Income"]]
p_value_data.head()
```

```python
poverty_m = p_value_data["Household Income"].mean()
poverty_m
```

```python
poverty_range = p_value_data["Household Income"].unique().tolist()
poverty_range
```

## Middle class range 

```python
middle_class = p_value_data.loc[p_value_data["Household Income"] >= 19000 ]
middle_class = middle_class.loc[middle_class["Household Income"] < 50000]
middle_class.sort_values(by = ["Household Income"], ascending = False)
```

```python
middle_class_mean = middle_class["Household Income"].mean()
middle_class_mean
```

## Upper middle class range

```python
upper_middle = p_value_data.loc[p_value_data["Household Income"] >= 50000]
upper_middle = upper_middle.loc[upper_middle["Household Income"] < 90000 ]
upper_middle.sort_values(by=["Household Income"], ascending = False)
```

```python
upper_middle_mean = upper_middle["Household Income"].mean()
upper_middle_mean
```

## High income class range

```python
high_income = p_value_data.loc[p_value_data["Household Income"] >= 90000]
high_income = high_income.loc[high_income["Household Income"] < 130000]
high_income.sort_values(by = ["Household Income"], ascending = False)
```

```python
high_income_mean = high_income["Household Income"].mean()
high_income_mean
```

```python
print(f' The average number of good range is {middle_class_mean}.')
print(f' The average number of medium range is {upper_middle_mean}.')
print(f' The average number of bad range is {high_income_mean}.')
```

```python
expected_value = (middle_class_mean + upper_middle_mean + high_income_mean / 3)
expected_value
```

```python
observed = pd.Series([high_income_mean, upper_middle_mean, middle_class_mean], index = ["Good", "Medium", "Bad"])
observed
```

```python
df = pd.DataFrame([observed]).T
df.head()
```

```python
df[1] = expected_value
```

```python
df.columns = ["Observed", "Expected"]
df
```

```python
# The degree of freedom is 3-1 = 2
# With a p-value of 0.05, the confidence level is 1.00-0.05 = 0.95.
critical_value = stats.chi2.ppf(q = 0.95, df = 2)
critical_value
```

```python
# Run the chi square test with stats.chisquare()
stats.chisquare(df['Observed'], df['Expected'])
```

### There is no significance between income and gun violence 
We conclude from p-value which is more than 0.05 there is no significant between income and gun crime.
